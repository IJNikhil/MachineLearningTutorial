```python
# Joins
# Create table from dictionary

Student_Data = {'Std_id' : [1, 2, 3, 4, 5], 'FName' : ['A', 'B', 'C', 'D', 'E'],
               'LName' : ['v', 'w', 'x', 'y', 'z']}
std = pd.DataFrame(Student_Data, columns = ['Std_id', 'FName', 'LName'])
print(std)
```

```python
# Single Bracket - series || Double Bracket - Data Frame
print("\n", std['FName'])
print("\n", type(std['FName']))
print("\n", std[['FName']])
print("\n", type(std[['FName']]))
```

```python
# to extrct(1st row 2nd column)

print(std.iloc[0:2, 0:2])
```

```python
# Extract with loc
print(std.loc[0:2, 'FName' : 'LName'])
```

```python
# Second table from dictionary
Student_Course = {'Stu_id' : [1, 2, 3, 4, 5], 
                'course' : ['ML', 'AI', 'Java', 'Python', 'CS'],
               'duration' : [3, 4, 5, 2, 1]}
course = pd.DataFrame(Student_Course)
print(course)
```

```python
# concat both tables/ Union
print(pd.concat([std, course]))
```

```python
# Axis = 0 - Union

print(pd.concat([std, course], axis = 0))
```

```python
# Index join
print(pd.concat([std, course], axis = 1))
```

```python
# Merge to join two tables
print(pd.merge(std, course, left_on = 'Std_id', right_on = 'Stu_id', how = 'inner'))
```

```python
# Left outer
print(pd.merge(std, course, left_on = 'Std_id', right_on = 'Stu_id', how = 'left'))
```

```python
# full outer
print(pd.merge(std, course, left_on = 'Std_id', right_on = 'Stu_id', how = 'outer'))
```

```python
# Que. 1) Write a program which will find factors of given number
#  and find whether the factor is even or odd.
# Hint: Use loop with if-else statements

# Input: 
# Number: 50

# Output:

# factor is odd:  1
# Factor is even: 2
# factor is odd:  5
# Factor is even: 10
# factor is odd:  25
# Factor is even: 50


num = 50
for i in range(1, num+1):
    rem = num%i
    
    if(rem == 0):
        if(i%2 == 0):
            print("Factor is even:", i)
        else:
            print("Factor is odd:", i)
            
    else:
        pass
    
```

```python
# Que. 3) Write a program which will find all the numbers between 1000 and
# 3000 (both included) such that each digit of a number is an even number.
# The numbers obtained should be printed in comma separated sequence on a
# Single line.

# Hint: In case of input data being supplied to the question, it should be
# assumed to be a console input. Divide each digit with 2 and verify is it
# even or not.

# output
# 2000,2002,2004,2006,2008,2020,2022,2024,2026,2028,2040,2042,2044,2046,
# 2048,2060,2062,2064,2066,2068,2080,2082,2084,2086,2088,2200,2202,2204,
# 2206,2208,2220,2222,2224,2226,2228,2240,2242,2244,2246,2248,2260,2262,
# 2264,2266,2268,2280,2282,2284,2286,2288,2400,2402,2404,2406,2408,2420,
# 2422,2424,2426,2428,2440,2442,2444,2446,2448,2460,2462,2464,2466,2468,
# 2480,2482,2484,2486,2488,2600,2602,2604,2606,2608,2620,2622,2624,2626,
# 2628,2640,2642,2644,2646,2648,2660,2662,2664,2666,2668,2680,2682,2684,
# 2686,2688,2800,2802,2804,2806,2808,2820,2822,2824,2826,2828,2840,2842,
# 2844,2846,2848,2860,2862,2864,2866,2868,2880,2882,2884,2886,2888


values = []
for i in range(1000, 3001):
    s = str(i)
    if(int (s[0])%2 == 0) and (int(s[1])%2 == 0) and (int(s[2])%2 == 0) and (int(s[3])%2 == 0):
        values.append(s)
    print(", ".join(values))
```

```python
# Que. 4) Write a program that accepts a sentence and calculate the number
# of letters and digits

# Hint: Use built-In function of string, -isdigit() and islpha()

# Input: 

# Suppose if the entered string is: Python0325

# Output:

# Then the output will be:

# LETTERS: 6

# DIGITS:4


s = "aac34520"
d = {"DIGITS" : 0, "LETTERS" : 0}
for c in s:
    if c.isdigit():
        d["DIGITS"] += 1
    elif c.isalpha():
        d["LETTERS"] += 1
    else:
        pass
print("LETTERS", d["LETTERS"])
print("DIGITS", d["DIGITS"])
```

```python
# Que. 5) Design a code which will find the given number is Palindrome number
# or not

# Hint: Use built-In function of string, reversed() (Not case Sensitive)

# The reversed() function returns a reversed iterator object.

# Example 

alph = ["a", "b", "c", "d"]
ralph = reversed(alph)
for x in ralph:
    print(x)
```

```python
# Input:  
# my_str = 'aIbohPhoBiA'
  
# Output:
# It is palindrome

# Change this value for a different output
my_str = 'aIbohPhoBiA'
# make it suitable for caseless comparison
my_str = my_str.casefold()
# reverse the string
rev_str = reversed(my_str)
# check if the string is equal to it's reverse
if list(my_str) == list(rev_str):
    print("It is palindrome")
else:
    print("It is not palindrome")
    
```

```python
# Option 2
my_str = 'aIbohPhoBiA'
my_str = my_str.casefold()
if my_str == my_str[::-1]:
    print("It is palindrome")
else:
    print("It is not palindrome")
```
