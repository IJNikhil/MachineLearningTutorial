```python
# Example 2, b = color 0 -circle

def f(t):
    return (np.exp(-t)*np.cos(2*np.pi*t))

t1 = np.arange(10, 20, 2)
t2 = np.arange(11, 21, 2)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2))
plt.title('Graph 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2))
plt.title('Graph 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.show()
```

```python
# tight_layout()

x1 = np.arange(10, 30, 2)
y1 = [1, 5, 6, 7, 8, 3, 4, 6, 5, 4]

x2 = np.arange(10, 30, 2)
y2 = [5, 4, 2, 8, 4, 6, 3, 4, 5, 4]

plt.subplot(211)
plt.plot(x1, y1)
plt.title('Graph 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.subplot(212)
plt.plot(x2, y2)
plt.title('Graph 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.tight_layout(pad = 1, h_pad = 1, w_pad = 1)

plt.show()
```

```python
# Que. 4) Please write a program which accepts a string from console and print the characters that have even indexes
# Input Example :
# H1e2l3l4o5w6o7r8l9d

# output is :
# Helloworld

s = input("Enter String like H1e2l3l4o5w6o7r8l9d means alphanumeric string which will contain alternate alphabet and numbers : \t")
s = s[::2]
print(s)
```

```python
# Que. 5) Please write a program which accepts a string from console and print it in reverse order
# Input Example :
# rise to vote sir

# output is :
# ris etov ot esir

s = input("Enter string which yout want to reverse:\t")

s = s[::-1]
print(s)
```

```python
# Que. 6) Please write a program which count and print the numbers of each character in a string input by a console
# Input Example :
# abcdefgabc

# output is :
# a, 2
# c, 2
# b. 2
# e, 1
# d, 1
# g, 1
# f, 1

dic = {}
s = input('Enter string : \t')
for s in s:
    dic[s] = dic.get(s, 0) + 1
    
print('\n'.join(['%s, %s'%(k, v) for k, v in dic.items()]))
    
```

```python
# Que. 9) Write a program to print the list after removing the 0th, 4th, 5th numbers in [12, 24, 35, 70, 120, 155]

li = [12, 24, 35, 70, 120, 155]
li = [x for (i, x) in enumerate(li) if i not in (0, 4, 5)]
print(li)
```

```python
# Que. 11) Write a program to compute 1/2+2/3+3/4+..+n/n+1 with a given n input by console (n>0)

# Example:
# input is 5
# output is: 
# 3.55 = 0.5+0.666+0.75+0.8+0.83

n = int(input("Enter a number :\t"))
sum = 0.0
for i in range(1, n+1):
    sum += float(float(i)/ (i+1))
print(sum)
```
