```python
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
import numpy as np
import pandas as pd
rcParams['figure.figsize'] = 15, 6
```

```python
# Multiple Graph
# Example 1

x1 = np.arange(10, 30, 2)
y1 = [1, 5, 6, 7, 8, 3, 4, 6, 5, 4]

x2 = np.arange(10, 30, 2)
y2 = [5, 4, 2, 8, 4, 6, 3, 4, 5, 4]

plt.subplot(211)
plt.plot(x1, y1)
plt.title('Graph 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.subplot(212)
plt.plot(x2, y2)
plt.title('Graph 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.show()
```

```python
# Example 2, b = color o -circle

def f(t):
    return (np.exp(-t)*np.cos(2*np.pi*t))

t1=np.arange(10,20,2)
t2= np.arange (11,21,2)
plt.subplot (211)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.title ('Graph 1')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')
plt.grid(color = 'k')


plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.title ('Graph 2')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')
plt.grid(color = 'k')
plt.show()
```

```python
# tight_layout()

x1 = np.arange(10, 30, 2)
y1 = [1, 5, 6, 7, 8, 3, 4, 6, 5, 4]

x2 = np.arange(10, 30, 2)
y2 = [5, 4, 2, 8, 4, 6, 3, 4, 5, 4]

plt.subplot(211)
plt.plot(x1, y1)
plt.title('Graph 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.subplot(212)
plt.plot(x2, y2)
plt.title('Graph 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color = 'k')

plt.tight_layout(pad = 1, h_pad = 1, w_pad = 1)

plt.show()
```

```python
# Create Graph from data set

plt.style.use('seaborn')

data = pd.read_csv('C:\\Users\\vijay\\Downloads\\MachineLeaninig\\Day_6\\Matplotlib_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label = 'Python')
plt.plot(ages, js_salaries, label = 'JavaScript')

plt.plot(ages, dev_salaries, color = '#444444', linestyle = '-', label = 'All Devs')
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()

plt.savefig ('Matplotlib_salary.png')

plt.show()
```

```python
# Create Subplot with figure and Axis

plt.style.use('seaborn')

data = pd.read_csv('C:\\Users\\vijay\\Downloads\\MachineLeaninig\\Day_6\\Matplotlib_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
print('Axis 1', ax1)
print('Axis 2', ax2)
ax1.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')

ax2.plot(ages, py_salaries, label = 'Python')
ax2.plot(ages, js_salaries, label = 'JavaScript')

ax1.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.grid(color = 'lightblue')
plt.show()

fig1.savefig('figure1.png')
fig2.savefig('figure2.png')
```

```python
# Save Graph
x= np.arange (0,100,2)
plt.plot(x, [x1 for x1 in x], label = 'Linear Graph')
plt.plot(x, [x1**2 for x1 in x], label = 'Square Graph')
plt.plot(x, [x1**3 for x1 in x], label = 'Three times of x')
plt.xlim ([0,20])
plt.ylim ([0,80])
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')
plt.legend()
plt.savefig ('Multi-Line-Graph.png')
plt.show()
```
